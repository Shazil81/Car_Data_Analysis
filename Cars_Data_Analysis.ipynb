# Data Analysis and Visualization

```python
# 1. Total No. Of Cars
print('Total Cars:', len(cars))
```

```python
# 2. Total No. Of Brands
print('Total Brands:', cars['Brand'].nunique())
```

```python
# 3. Avg Price of Cars
print('Average Price:', cars['Price'].mean())
```

```python
# 4. Avg Mileage
print('Average Mileage:', cars['Mileage'].mean())
```

```python
# 5. Count of Cars by Fuel Type
fuel_type_count = cars['Fuel_Type'].value_counts()
plt.figure(figsize = (8, 5))
ax = fuel_type_count.plot(kind = 'bar', color = 'skyblue')

plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Count of Cars by Fuel Type')

ax.bar_label(ax.containers[0])
plt.show()
```

```python
# 6. Brands with Avg Price
avg_price_brand = cars.groupby('Brand')['Price'].mean().reset_index().sort_values(by = 'Price')
ax = avg_price_brand.plot(kind = 'bar', x = 'Brand', legend = False, figsize = (8, 5), color = 'orange')

plt.ylabel('Average Price')
plt.title('Average Price by Brand')

ax.bar_label(ax.containers[0], fontsize = 6)
plt.tight_layout()
plt.show()
```

```python
# 7. Brand with Highest Avg Price
highest_avg_price_brand = cars.groupby('Brand')['Price'].mean().reset_index().sort_values(by = 'Price', ascending = False).head(1)
highest_avg_price_brand
```

```python
# 7. Best Fuel Type by Avg Mileage
best_fuel_avg_Mileage = cars.groupby('Fuel_Type')['Mileage'].mean().reset_index().sort_values(by = 'Mileage', ascending = False)
ax = best_fuel_avg_Mileage.plot(kind = 'bar', x = 'Fuel_Type', legend = False, figsize = (8, 5), color = 'green')
plt.ylabel('Average Mileage')
plt.title('Average Mileage by Fuel Type')

ax.bar_label(ax.containers[0])
plt.tight_layout()
plt.show()
```

```python
# 8. Average Price per Brand for Diesel Cars
avg_price_brand = cars[cars['Fuel_Type'] == 'Diesel'].groupby('Brand')['Price'].mean().reset_index().sort_values(by = 'Price')
ax = avg_price_brand.plot(kind = 'bar', x = 'Brand', legend = False, figsize = (8, 5), color = 'red')

plt.ylabel('Average Price')
plt.title('Average Price by Brand for Diesel Cars')

ax.bar_label(ax.containers[0], fontsize = 6)
plt.tight_layout()
plt.show()
```

```python
# 9. Total No of Automatic and Manual Cars by Brand
trans_count = cars.groupby('Brand')['Transmission'].value_counts().unstack(fill_value = 0).reset_index()
ax = trans_count.plot(kind = 'bar', stacked = True, x = 'Brand', figsize = (9, 6))

plt.ylabel('Number of Cars')
plt.title('Transmission Type Distribution by Brand')
plt.legend(title = 'Transmission', fontsize = 6)

for container in ax.containers:
    ax.bar_label(container, label_type = 'center', fontsize = 8)
    
plt.tight_layout()
plt.show()
```

```python
# 10. Avg Price Difference Between Automatic & Manual Cars
avg_auto_price = cars[cars['Transmission'] == 'Automatic']['Price'].mean()
avg_manual_price = cars[cars['Transmission'] == 'Manual']['Price'].mean()
price_difference = avg_auto_price - avg_manual_price
print('Price Difference :', price_difference)
```

```python
# 11. Top Brands by Price per CC
cars['PricePerCC'] = cars['Price'] / cars['Engine_CC']
top_brands_pricepercc = cars.groupby('Brand')['PricePerCC'].mean().reset_index().sort_values(by = 'PricePerCC', ascending = False).head(5)
ax = top_brands_pricepercc.plot(kind = 'barh', x = 'Brand', legend = False, figsize = (10,5), color = 'purple')

plt.ylabel('Price Per CC')
plt.title('Top Brands by Price per CC')
ax.bar_label(ax.containers[0])
plt.tight_layout()
plt.show()
```

```python
# 12. Top 3 Most Expensive Cars Per Brand
top3_expensive = cars[cars.groupby('Brand')['Price'].rank(method = 'first', ascending = False) <= 3]
top3_expensive[['Brand', 'Model', 'Year', 'Price', 'Fuel_Type']]
```

```python
# 13. Top 3 Least Expensive Cars Per Brand
top3_least_expensive = cars[cars.groupby('Brand')['Price'].rank(method = 'first', ascending = True) <= 3]
top3_least_expensive[['Brand', 'Model', 'Year', 'Price', 'Fuel_Type']]
```

```python
# 14. Brand Performance Percentile by Service Cost
brand_service_cost = cars.groupby('Brand')['Service_Cost'].mean().reset_index()
brand_service_cost['Quartile'] = pd.qcut(brand_service_cost['Service_Cost'], 4, labels = False) + 1

sns.set(style = 'white')
plt.figure(figsize = (10, 6))
ax = sns.barplot(data = brand_service_cost, x = 'Brand', y = 'Service_Cost', hue = 'Quartile', dodge = False, palette = 'Set2')
plt.xticks(rotation = 45)

for i in ax.containers:
    ax.bar_label(i, label_type = 'center')

plt.title("Avg Service Cost by Brand with Quartile")
plt.tight_layout()
plt.show()
```

```python
# 15. Brands with Above-Average Mileage Compared to All Brands
overall_avg_mileage = cars['Mileage'].mean()
brands_above_avg_mileage = cars.groupby('Brand')['Mileage'].mean().reset_index()
brands_above_avg_mileage = brands_above_avg_mileage[brands_above_avg_mileage['Mileage'] > overall_avg_mileage]
brands_above_avg_mileage
```

```python
# 16. Outlier Cars within the Brand
plt.figure(figsize=(12, 6))
sns.boxplot(data = cars, x = 'Brand', y = 'Price')
plt.xticks(rotation = 45, ha = 'right')
plt.title('Price Distribution with Outliers by Brand')
plt.tight_layout()
plt.show()
```

```python
# 17. Car Models That are Top 1 in Price in Each Year
cars['rank'] = cars.groupby('Year')['Price'].rank(method = 'dense', ascending = False)
top1_price_models_each_year = cars[cars['rank'] == 1]
top1_price_models_each_year[['Year', 'Model', 'Price']]
```
